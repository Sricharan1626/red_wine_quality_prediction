```python
import pandas as pd
```

```python
df= pd.read_csv("/content/winequality-red.csv")
```

```python
df.head()
```

```python
#Checking any null values

any_null=df.isnull().sum()
print(any_null[any_null>0])

#NO null values it is clean data
X=df.drop('quality',axis=1)
y=df.quality
```

```python
#splitiing of data

from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X,y, train_size =0.8, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

model = RandomForestRegressor(n_estimators=1000, random_state=1)

Train_model=model.fit(train_X,train_y)
prediction = Train_model.predict(test_X)
prediction = [round(num) for num in prediction]
mae=mean_absolute_error(test_y,prediction)
print(prediction)
```

```python
test_y.head()
```

```python
print(mae)
```

```python
#Still need more accuracy so lets train whole dataset

full_train = model.fit(X,y)
```

```python
prediction = full_train.predict(test_X)
print(test_y.head())

prediction = [round(num) for num in prediction]
print(prediction)
mae = mean_absolute_error(test_y,prediction)
score = r2_score(test_y,prediction)
```
